```r
#importing package Data Visualization
library(ggplot2)
```

```r
ggplot(df_House_Price, aes(x = GarageArea, y = SalePrice)) +
    geom_point()
```

```r
#Plot without grouping based on Garage Type
ggplot(df_House_Price, aes(x = GarageArea, y = SalePrice)) +
    geom_point(aes(color = factor(GarageCond)))
```
